## Transformación

Manos a la obra, a limpiar datos ;) 

Si la columna de edad tiene valores nulos , vamos a rellenar esos valores nulos por el promedio de edad según el género. 

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df.isnull().sum()
```

Ahora eliminemos todos los registros que aún tengan datos nulos en cualquier columna

```python
df = df.dropna()
df
```

Elimina todos los registros duplicados

```python
df.drop_duplicates()
```

Cambiemos todos los datos con valor tipo string a minúscula

```python
df = df.astype(str).applymap(lambda x: x.lower())
df
```

Los correos están horribles y no nos servirán de mucho, eliminemoslos ;) 

```python
df = df.drop(columns= ['email'])
```

Cambia el orden de las columnas, primero apellido y luego nombre

```python
df = df[['id', 'last_name', 'first_name', 'gender', 'country', 'age']]
df.head(3)
```

Por último , asegúrate que la edad esté en dato tipo integer

```python
df['age'] = df['age'].astype(float).astype(int)
df.dtypes['age']



```

## Hora de insights... 

¿Cuántas personas son de España? retorna en dato tipo entero 

```python
origen_espana = 0
for country in df['country']:
    if country == 'España':
        origen_espana += 1
origen_espana
```

```python
#Solucion 2:
origen_espana = (df['country'] == 'España').sum()
origen_espana
```

¿De cuántos países diferentes son los clientes? retorna dato de tipo entero 

```python
df.country.unique().size
```

¿Cuál es el país con mayor cantidad de clientes? 

```python
moda = df['country'].mode().values[0]
moda
```

Ordenando los países por orden alfabético , cuál es el último país?

```python
df['country'] = df['country'].sort_values(ascending=False)
```

Cuál es el nombre más común de hombre y cuál de mujer ? 

```python
nombres_hombres = df[df['gender'] == 'male']['first_name']
nombres_mujeres = df[df['gender'] == 'female']['first_name']
nombre_mas_comun_hombres = nombres_hombres.mode().values[0]
nombre_mas_comun_mujeres = nombres_mujeres.mode().values[0]
print(f'El nombre más común en hombres es {nombre_mas_comun_hombres} y el nombre más comun mujeres es {nombre_mas_comun_mujeres}')
```

Qué país tiene el promedio de edad más alto de mujeres? 

```python
promedio_mujeres_pais = df[df['gender'] == 'female'].groupby('country')['age'].mean()
pais_con_promedio_mas_alto = promedio_mujeres_pais.idxmax()
pais_con_promedio_mas_alto

```

Qué país tiene el promedio de edad más bajo de hombres ? 

```python
promedio_hombres_pais = df[df['gender'] == 'male'].groupby('country')['age'].mean()
pais_con_promedio_mas_bajo = promedio_hombres_pais.idxmin()
pais_con_promedio_mas_bajo
```

Por último crea una nueva columna de edad dada en meses donde sus valores sea la edad de clientes dada en meses, suponiendo que todos tienen años recién cumplidos :D 

```python
df['age_months'] = df['age'] * 12
df
```
